```python
import matplotlib.pyplot as plt

xaxis = [elem/10 for elem in list(range(-100, 100))]
yaxis = [i for i in range(200)]
print(len(xaxis))
print(len(yaxis))
plt.bar(xaxis, yaxis)
plt.yscale('log')
plt.xticks([-10+i for i in range(21)])
plt.savefig('test1.png', dpi=400)
plt.yscale('linear')
plt.savefig('test2.png', dpi=400)
plt.cla()
plt.bar(xaxis, yaxis)
plt.savefig('test3.png', dpi=400)
```
