```python
import tqdm
from sen12mscr.base_dataset import SEN12MSCR_A
import numpy as np
import matplotlib.pyplot as plt
import rasterio

def load_s1(path, imgTransform):
    with rasterio.open(path) as data:
        s1 = data.read()
    s1 = s1.astype(np.float32)
    s1 = np.nan_to_num(s1)
    s1 = np.clip(s1, -25, 0)
    if not imgTransform:
        s1 /= 25
        s1 += 1
    s1 = s1.astype(np.float32)
    return s1

def false_rgb6(s1, idx):
    img = np.clip(s1, -25, 0)
    
    result = np.divide(img[0], img[1], out=np.zeros_like(img[0]), where=(img[1] != 0))
    
    img[0] /= 25
    img[0] += 1
    img[1] /= 25
    img[1] += 1
    result = (result-np.min(result))/(np.max(result)-np.min(result))    
    
    false_rgb = np.stack([img[0], img[1], result], axis=-1)
    plt.imsave(f'SEN12MS-CR-rgb/false_rgb_{idx:03}.png', false_rgb, cmap='gray')

def false_rgb_new(s1, idx):
    img = np.clip(s1, -25, 0)
    
    result = np.divide(img[0], img[1], out=np.zeros_like(img[0]), where=(img[1] != 0))
    
    img[0] = (img[0]-np.min(img[0]))/(np.max(img[0])-np.min(img[0])) 
    img[1] = (img[1]-np.min(img[1]))/(np.max(img[1])-np.min(img[1])) 
    result = (result-np.min(result))/(np.max(result)-np.min(result))    
    
    false_rgb = np.stack([img[0], img[1], result], axis=-1)
    plt.imsave(f'SEN12MS-CR-rgb-new/false_rgb_{idx:03}.png', false_rgb, cmap='gray')

dataset = SEN12MSCR_A('/media/khj/portable-hdd1/SEN12MSCR', s1_rgb_composite=None, s1_rescale_method=None)
# 
# for idx, s1 in tqdm.tqdm(enumerate(dataset)):
#     false_rgb6(s1['S1'], idx)
for idx, s1 in tqdm.tqdm(enumerate(dataset)):
    false_rgb_new(s1['S1'], idx)
```

```python
from sen12mscr.base_dataset import SEN12MSCR_AB
import matplotlib.pyplot as plt
import numpy as np
import os
import tqdm

def normalize(img):
    img = img.astype(np.float64)
    img -= np.mean(img)
    img_std = np.std(img)
    img += img_std
    img /= img_std * 4.0
    img = np.clip(img, 0, 1)
    return img

loader = SEN12MSCR_AB(    root='/home/khj/SEN12MSCR', split='test_use_all', season='all', s1_rescale_method='none', s2_rescale_method='none', s1_rgb_composite='add')

logdir = '/home/khj/sen12ms-log'

for sample in tqdm.tqdm(loader):
    path = sample['B']['S2 path']
    imgname = os.path.splitext(os.path.basename(path))[0]+'.png'
    dirname = os.path.basename(os.path.dirname(path))
    basename = os.path.basename(os.path.dirname(os.path.dirname(path)))
        
    img = sample['B']['S2']
    img = normalize(img)
    img = np.transpose(img, (1, 2, 0))
    
    if not os.path.isdir(logdir+'/'+dirname):
        os.makedirs(logdir+'/'+dirname)
    
    path = logdir+'/'+dirname+'/'+imgname
    plt.imsave(path, img)
    
```
