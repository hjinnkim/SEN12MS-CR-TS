```python
def normalize(img):
    img = img.astype(np.float64)
    img -= np.mean(img)
    img_std = np.std(img)
    img += img_std
    img /= img_std * 4.0
    img = np.clip(img, 0, 1)
    return img
```

```python
a = normalize(np.flip(sen12mscr[100000]['target']['S2'][1:4].transpose(1, 2, 0), axis=2))
b = normalize(sen12mscr[100000]['target']['S2'][1:4].transpose(1, 2, 0))
```
